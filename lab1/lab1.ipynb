```python
import numpy as np
import pandas as pd
import time
```

```python
class Graph:
    def __init__(self, vertices, weights):
        self.vertices = vertices
        self.weights = weights
        self.vertex_num = len(vertices)
```

```python
class MinimumSpanningTree:
    def __init__(self, graph):
        self.graph = graph

    def prim(self, start):
        visited = [False] * self.graph.vertex_num
        visited[start] = True
        vertices = []
        summ = 0

        for k in range(1, self.graph.vertex_num):
            min_weight = 1e8
            i1 = -1
            i2 = -1
            for i in range(self.graph.vertex_num):
                if not visited[i]:
                    continue
                for j in range(self.graph.vertex_num):
                    if (self.graph.weights[i][j] < min_weight) & (not visited[j]):
                        min_weight = self.graph.weights[i][j]
                        i1 = i
                        i2 = j
            visited[i2] = True
            summ += min_weight
            vertices.append([self.graph.vertices[i1], i1, self.graph.vertices[i2], i2])
        return vertices, summ
```

```python
start = time.time()

coords = pd.read_excel('Table.cities.coord.xls')

weights = np.array(coords.iloc[1:, 3:])
weights = np.where(weights == ' ', 1e8, weights)
weights = weights.astype('float')

cities = coords.iloc[0, 3:]
cities = np.array(cities)

graph = Graph(cities, weights)
tree = MinimumSpanningTree(graph)
edges, summ = tree.prim(0)

end = time.time()

print(f'Work time {end - start} s')
print(f'Sum length {summ}')
for edge in edges:
    print(f'{edge[0]} ({edge[1]}) - {edge[2]} ({edge[3]})')
```
